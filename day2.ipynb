# Adding Removing Sorting elements

#### np.sort() np.concatenate()

```python
import numpy as np
```

```python
arr = np.array([2,1,3,4,5,6,7,10,3])
```

```python
print(f"dimension = {arr.ndim}")
print(f"shape = {arr.shape}")
print(f"size = {arr.size}")
print(f"dtype = {arr.dtype}")
```

```python
np.sort(arr)
```

```python
a = np.array([1,2,3,4])
b = np.array([5,6,7,8])
np.concatenate((a,b))
```

##### axis = 0 -> Rows; axis = 1 -> Columns

```python
a = np.array([
    [1,2,3],
    [4,5,6]
])
b = np.array([
    [1,5,9],
    [2,4,8]
])
print("Along Colummns")
print(np.concatenate((a,b),axis=1))
print("Along Rows")
print(np.concatenate((a,b),axis=0))
```

## Indexing and Slicing

```python
arr = np.arange(4,10,2)
arr
```

```python
print(arr[1:])
print(arr[-2:])
```

```python
arr = np.array([
    [
        [1,2,3],
        [2,3,4]
    ],
    [
        [4,5,6],
        [5,6,7]
    ],
    [
        [9,8,7],
        [8,5,2]
    ]
])
print(arr)
```

```python
print(arr.ndim)
print(arr.shape)
print(arr.size)
print(arr.dtype)
```

```python
print(arr[arr % 2 == 0])
print(arr[arr < 5])
print(arr[(arr > 5) & (arr < 8)])
```

```python
arr1 = np.array([[1,2,3],[3,4,5]])
arr2 = np.array([[4,5,6],[2,4,6]])
print(f"addition\n{arr1 + arr2}")
print(f"subtraction\n{arr1 - arr2}")
print(f"Multiplication\n{arr1 * arr2}")
print(f"division\n{arr1 / arr2}")
```

##### Vector and Scalar

```python
print(arr1 * 2)
print(arr1 + 10)
```

#### min max sum mean prod std

```python
arr = np.array([1,2,4])
print(arr.max())
print(arr.min())
print(arr.sum())
```

#### Transposing Reshaping a matrix

```python
arr = np.array([
    [1,2,3],
    [4,5,6],
    [7,5,3]
])
print(arr)
print(arr.T)
print(arr.transpose())
```

```python
arr = np.array([1,2,3,4,5,6])
print(np.flip(arr))
arr = np.array([
    [1,2,3],
    [2,3,4],
    [3,4,5]
])
print(arr)
print(np.flip(arr))
```
