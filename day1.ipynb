## Numpy (BASICS)

```python
import numpy as np
```

```python
a = np.array([
    [1,2,3],
    [2,4,6]
])
a.shape
```

#### Restrictions
1) All elements of the array must be of the same type of data.
2) Once created, the total size of the array can't change.

```python
# column is at the last and row is at the second last
a[1,1] 
```

1) 0-D -> Scaler
2) 1-D -> Vector
3) 2-D -> Matrix
4) N-D -> Tensors

#### Array attributes

##### ndim, shape, size, dtype

```python
# The number of dimensions of an array is contained in the ndim attribute
a.ndim
```

```python
# The shape of an array is a tupple of non-negative integers that specify the number of elements along each dimension.
a.shape
```

```python
# The fixed, total number of elements in array is contained in the size attribute.
a.size
```

```python
# Arrays are typically “homogeneous”, meaning that they contain elements of only one “data type”. The data type is recorded in the dtype attribute.
a.dtype
```

#### Basic Array Methods
##### np.zeros(), np.ones(), np.empty(), np.arange(), np.linspace()

```python
# Creating an array filled with zeroes
b = np.zeros(4)
b
```

```python
# Creating an array filled wih one's
print(np.ones(3))
print(np.ones(3, dtype=np.int64))
```

```python
# Creating an array with empty elements
d = np.empty(2)
d
```

```python
# creating an array with a range of elements
print(np.arange(4)) # 0-3 (start, stop, step)
print(np.arange(2,6,2))
```

```python
# Create an array with values that are spaced linearly in a specified interval
np.linspace(0,10,num=5)
```
